4. Пусть в некотором векторном пространстве сущность «Университет ИТМО» представлена в виде вектора `[0.52, 0.8, 0.95, 0.29, 0.76]`, а предикат «находится в городе» `[0.39, 0.74, 0.26, 0.5, 0.32]`. Подсчитайте векторное представление объекта триплета «Университет ИТМО — находится в городе — Санкт-Петербург».

```python
from numpy import array

subject = array([0.52, 0.8, 0.95, 0.29, 0.76])
predicate = array([0.39, 0.74, 0.26, 0.5, 0.32])
print(str(subject + predicate).replace(' ', ','))
```
